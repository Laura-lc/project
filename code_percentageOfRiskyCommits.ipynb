```python
# Percentage of Risky Commits
import csv
import os
import pandas as pd
import time
import datetime
from numpy import *

csv.field_size_limit(100000000)
for info in os.listdir(path_to_the_returned_csv_files):
    domain = os.path.abspath(r'path_to_the_returned_csv_files')   
    # e.g. 1001_AbFab3D-0ca2ced9-43c4-41e6-ad04-83ac295eabce.csv
    # to get the projectID
    fileNameSplit = info.split('_')
    whole_path = os.path.join(domain,info)    
    column = []   
    with open(whole_path,'r',encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        column = [row['author_date'] for row in reader]
    interval_start = 0
    interval_end = 0
    c1 = 0
    
    for index, val in enumerate(column):
        split_date = val.split()
        if split_date[1] == 'Sep':
            split_date[1] = '09'
        if split_date[1] == 'Oct':
            split_date[1] = '10'
        if split_date[1] == 'Nov':
            split_date[1] = '11'
        if split_date[1] == 'Dec':
            split_date[1] = '12'
        if split_date[1] == 'Jan':
            split_date[1] = '01'
        if split_date[1] == 'Feb':
            split_date[1] = '02'
        if split_date[1] == 'Mar':
            split_date[1] = '03'            
        if split_date[1] == 'Aug':
            split_date[1] = '08'
        if split_date[1] == 'Jul':
            split_date[1] = '07'
        if split_date[1] == 'Jun':
            split_date[1] = '06'
        if split_date[1] == 'May':
            split_date[1] = '05'    
        if split_date[1] == 'Apr':
            split_date[1] = '04'  
            
        # convert 2019-03-3 to 2019-03-03
        single_num = ['1','2','3','4','5','6','7','8','9']
        if split_date[2] in single_num:
            split_date[2] = '0' + split_date[2]
            
        # date format
        format_date = split_date[4] + '-' + split_date[1] + '-' + split_date[2]
        # time period setting
        if c1 == 0 and format_date < '2020-03-13':
            c1 += 1
            interval_start = index        
        if format_date < '2019-09-13':
            interval_end = index
            break
            
    contain_bugs = []    
    risky_commit_count = 0
    risky_commit_percentage = 0
    total_commit = 0    
    with open(whole_path,'r',encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        contain_bugs = [row['contains_bug'] for row in reader]
        contain_bugs = contain_bugs[interval_start:interval_end]
        
        for x in contain_bugs:
            if x == 'True' or x == 'TRUE':
                risky_commit_count += 1                
        total_commit = len(contain_bugs)
        if total_commit != 0:
            risky_commit_percentage = risky_commit_count/total_commit
        else:
            risky_commit_percentage = '----'        
    print(fileNameSplit[0],total_commit,'{:.2%}'.format(risky_commit_percentage))
```
