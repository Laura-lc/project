```python
# Distribution of Modified Code - entropy
import csv
import os
import pandas as pd
import time
import datetime
from numpy import *
csv.field_size_limit(100000000)
import statistics 
for info in os.listdir(path_to_the_returned_CSV_files):
    domain = os.path.abspath(r'path_to_the_returned_CSV_files')
    # e.g. 1001_AbFab3D-0ca2ced9-43c4-41e6-ad04-83ac295eabce.csv
    fileNameSplit = info.split('_')
    whole_path = os.path.join(domain,info)    
    column = [] 
    modified_files = []
    modified_files_sum = 0
    
    with open(whole_path,'r',encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        column = [row['author_date'] for row in reader]
        
    interval_start = 0
    interval_end = 0
    c1 = 0   
    for index, val in enumerate(column):
        split_date = val.split()
        if split_date[1] == 'Sep':
            split_date[1] = '09'
        if split_date[1] == 'Oct':
            split_date[1] = '10'
        if split_date[1] == 'Nov':
            split_date[1] = '11'
        if split_date[1] == 'Dec':
            split_date[1] = '12'
        if split_date[1] == 'Jan':
            split_date[1] = '01'
        if split_date[1] == 'Feb':
            split_date[1] = '02'
        if split_date[1] == 'Mar':
            split_date[1] = '03'
        if split_date[1] == 'Aug':
            split_date[1] = '08'
        if split_date[1] == 'Jul':
            split_date[1] = '07'
        if split_date[1] == 'Jun':
            split_date[1] = '06'
        if split_date[1] == 'May':
            split_date[1] = '05'    
        if split_date[1] == 'Apr':
            split_date[1] = '04'  
               
        # convert 2019-03-3 to 2019-03-03
        single_num = ['1','2','3','4','5','6','7','8','9']
        if split_date[2] in single_num:
            split_date[2] = '0' + split_date[2]            
        # format
        format_date = split_date[4] + '-' + split_date[1] + '-' + split_date[2]
        if c1 == 0 and format_date < '2020-03-13':
            c1 += 1
            interval_start = index        
        if format_date < '2019-09-13':
            interval_end = index
            break
            
    average_entropy = 0
    with open(whole_path,'r',encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        commit_entropys = [row['entropy'] for row in reader]
        commit_entropys = commit_entropys[interval_start:interval_end]

        float_commit_entropys = []
        for x in commit_entropys:
            if x != '':
                float_commit_entropys.append(float(x))       
        # mean
        commit_entropys_sum = sum(float_commit_entropys)
        total_commit = len(float_commit_entropys)
        average_entropy = commit_entropys_sum/total_commit
        
        # median
        median = 0
        float_commit_entropys.sort()
        if total_commit % 2 == 0: 
            m1 = float_commit_entropys[total_commit//2] 
            m2 = float_commit_entropys[total_commit//2 - 1] 
            median = (m1 + m2)/2
        else: 
            median = float_commit_entropys[total_commit//2]         
    print(fileNameSplit[0],total_commit,"%.2f" % average_entropy,"%.2f" % median)
```
