```python
# Function and variables replacement for Python projects
import re  
p1 = '[a-z][a-z1-9]+([A-Z][a-z1-9]+)+' # camelCase
p2 = '[A-Z][a-z1-9]+([A-Z][a-z1-9]+)+' # Pascal Case
p3 = '[a-z1-9]+([_][a-z1-9]+)+' # e.g. set_vasible_value                            
# function with parameter,e.g.setBackgroundMode(SWT.INHERIT_DEFAULT)
p6 = '([(](.+)[)])'   
inclined_rod = '/'
dotM = '.'
bracketsM = '('
bracketsMM = '()'
function_name_list=[] # function names list
variable_name_list=[] # variable names list

library_class_names = ['False','None','True']
fo = open(path_for_store,"w",encoding="utf-8")
for each_line in open(path_of_comments,'r'):
    y = each_line.split()
    for idx, i in enumerate(y):       
        v1 = re.match(p1,i,flags=0)
        v2 = re.match(p2,i,flags=0)
        v3 = re.match(p3,i,flags=0)

        if i.find(inclined_rod) == -1 and i.find(bracketsMM) == -1:  
            _func_ = 'func' in y
            _funcs_ = 'funcs' in y
            _function_ = 'function' in y
            _Function_ = 'Function' in y
            
            if dotM in i:
                i_lis = i.split('.')
                _one_ = i_lis[1] in library_class_names
                _zero_ = i_lis[0] in library_class_names
                # "." is not in the end
                if _one_ == False and i_lis[1] != "":                    
                    v11 = re.match(p1,i_lis[1],flags=0)
                    v22 = re.match(p2,i_lis[1],flags=0)
                    v33 = re.match(p3,i_lis[1],flags=0) 
                    if v11 != None or v22 != None or v33 != None:
                        function_name_list.append(i)
                        each_line = each_line.replace(i,'FUNCTION')
                # "."is a period
                else:
                    if _zero_ == False:
                        if v1 != None :
                            function_name_list.append(i)
                            each_line = each_line.replace(i,'FUNCTION')
                            
                        if v2 != None:
                            if _func_ == True or _funcs_ == True or _function_ == True or _Function_ == True:
                                function_name_list.append(i)
                                each_line = each_line.replace(i,'FUNCTION')
                        
                        if v3 != None:
                            if _func_ == True or _function_ == True:
                                function_name_list.append(i)
                                each_line = each_line.replace(i,'FUNCTION')
                        
                        if i == 'call':
                            if y[idx+1] != "":
                                v333 = re.match(p3,y[idx+1],flags=0)
                                if v333 != None:
                                    function_name_list.append(i)
                                    each_line = each_line.replace(i,'FUNCTION')
            # no "."
            else:
                if v1 != None :
                    function_name_list.append(i)
                    each_line = each_line.replace(i,'FUNCTION')
                    
                if v2 != None:
                    if _func_ == True or _funcs_ == True or _function_ == True or _Function_ == True:
                        function_name_list.append(i)
                        each_line = each_line.replace(i,'FUNCTION')

                if v3 != None:
                    if _func_ == True or _function_ == True:
                        function_name_list.append(i)
                        each_line = each_line.replace(i,'FUNCTION')
                    
                #judge by preposition
                if i == 'in' or i == 'for' or i == 'call':
                    if (idx+1) < len(y) and y[idx+1] != "":                     
                        v33 = re.match(p3,y[idx+1],flags=0)
                        if v33 != None:
                            function_name_list.append(y[idx+1])
                            each_line = each_line.replace(y[idx+1],'FUNCTION')
            
        # followed by a (), we say it a FUNCTION, like'init(),'       
        if "()" in i:
            function_name_list.append(i)
            each_line = each_line.replace(i,'FUNCTION')
                               
        # function with parameter, e.g.setBackgroundMode(SWT.INHERIT_DEFAULTï¼‰
        if bracketsM in i:
            split_list = i.split('(')
            if split_list[0] != "":               
                if v1 != None or v2 != None or v3 != None:
                    ret6 = re.search(p6,'(' + split_list[1])
                    if ret6 != None:
                        function_name_list.append(i)
                        each_line = each_line.replace(i,'FUNCTION')                        
    fo.write(each_line)
fo.close()
```
