```python
# Measure the percentage of unique comment excluding stopwords
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.tokenize import RegexpTokenizer
import os
import numpy as np

##convert a list to string 
def listToString(s):      
    str1 = " "     
    return (str1.join(s)) 

g = os.walk(r"path_to_your_projects")
# match_project_ID - i.e. a set of projectName : projectID in a txt file
# just for converting the project name to project ID
match_project_ID = 'predefine_projectId_for_each_project'

tokenizer = RegexpTokenizer(r'\w+')
stop_words = set(stopwords.words('english'))
print('proID TotalC duplicatePer')
for path,dir_list,file_list in g:  
    for dir_name in dir_list:
        proj_id = ''
        total_comment = 0
        comment_after_clean = []
        check_duplicated = []
        duplicated_percent = 0
        unique_percent = 0
        
        # matching the project id
        for each_line in open(match_project_ID,'r'):
            y = each_line.split()
            if dir_name and dir_name == y[0]:
                proj_id = y[1]          
        os.chdir(path+'/'+dir_name)
        # open the comment file of current project
        f = open('commitmessages.txt','r',encoding='UTF-8') 
        line = f.readline()
        
        while line:
            total_comment += 1
            # preprocessing
            result = tokenizer.tokenize(line)
            word_tokens = word_tokenize(listToString(result))
            filtered_sentence = [w for w in word_tokens if not w in stop_words]
            filtered_sentence = []
            
            for w in word_tokens: 
                if w not in stop_words: 
                    filtered_sentence.append(w) 
            comment_after_clean.append(listToString(filtered_sentence))
    
            for i in comment_after_clean:
                if i not in check_duplicated:
                    check_duplicated.append(i)            
            #duplicated_percent = 1 - len(check_duplicated)/len(comment_after_clean)
            unique_percent = len(check_duplicated)/len(comment_after_clean)            
            line = f.readline()                  
        f.close()
        print(proj_id,'   ',total_comment,'   ','{:.2%}'.format(duplicated_percent))
```
