```python
# Function and variables replacement for Java projects
import re  
p1 = '[a-z][a-z1-9]+([A-Z][a-z1-9]+)+' # camelCase，e.g. setThePath() 
p6 = '([(](.+)[)])'  # function with parameter,e.g. setBackgroundMode(SWT.INHERIT_DEFAULT)
p7 = '[a-z]([A-Z][a-z]+)+'  # e.g. ‘sProxyLock’ should be a variable
inclined_rod = '/'
dotM = '.'
bracketsM = '('
bracketsMM = '()'
function_name_list=[] # function names list
variable_name_list=[] # variable names list
# common library classes
library_class_names = ['StringBuffer ','StringBuilder ','GregorianCalendar ','Hashtable ','StringTokenizer','BitSet','HashSet','ArrayList',
                       'TypeScript','SqlStatements','WindowEvent','ValueOutputStream','TimeoutException','StringReader','StringIndexOutOfBoundsException',
                       'StringBufferInputStream','StringWriter','ArrayIndexOutOfBoundsException','ArrayType']
exclde_list = ['defaultRowFetchSize=1000','()','().']

fo = open(path_for_store,"w",encoding="utf-8")
for each_line in open(path_of_comments,'r',encoding="utf-8"):
    y = each_line.split()
    for idx, i in enumerate(y):       
        v1 = re.match(p1,i,flags=0)   
        v7 = re.match(p7,i,flags=0)  
        i.replace(' ','')       
        _yes_or_not = i in exclde_list
        
        #  not contain '/','()'
        if _yes_or_not == False and i.find(inclined_rod) == -1 and i.find(bracketsMM) == -1:                       
            if dotM in i:
                i_lis = i.split('.')
                # "." is not at the end
                if i_lis[1] != "": 
                    #library_name = i_lis[1] in library_class_names
                    v11 = re.match(p1,i_lis[1],flags=0)
                    v77 = re.match(p7,i_lis[1],flags=0)
                    # camelCase
                    if v11 != None:
                        function_name_list.append(i)
                        each_line = each_line.replace(i,'FUNCTION')
                    # variable
                    if v77 != None:
                        variable_name_list.append(i)
                        each_line = each_line.replace(i,'VARIABLE')
                # "." is period
                else:
                    # Functions, like getDataSources
                    if v1 != None :
                        function_name_list.append(i)
                        each_line = each_line.replace(i,'FUNCTION')
                    if v7 != None:
                        variable_name_list.append(i)
                        each_line = each_line.replace(i,'VARIABLE')
            # no "."
            else:
                # camelCase
                if v1 != None :
                    function_name_list.append(i)
                    each_line = each_line.replace(i,'FUNCTION')
                if v7 != None:
                    variable_name_list.append(i)
                    each_line = each_line.replace(i,'VARIABLE')
                
        # followed by a (), like，setBackgroundMode()    
        if "()" in i:
            function_name_list.append(i)
            each_line = each_line.replace(i,'FUNCTION')
            
        # function with parameter
        if bracketsM in i:
            split_list = i.split('(')
            ret6 = re.search(p6,'(' + split_list[1])
            if split_list[0] != "":
                if v1 != None and ret6 != None:
                    function_name_list.append(i) 
                    each_line = each_line.replace(i,'FUNCTION')         
    fo.write(each_line)
fo.close()
```
